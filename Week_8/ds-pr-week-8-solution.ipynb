### 6. Correlations

```python
# For numerical values
# Correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Map')
plt.show()
```

```python
# Confirm the correlation between SleepHours and StressLevel using Spearman and Pearson (coefficient and p-value)
from scipy.stats import pearsonr, spearmanr

pearson_corr, pearson_p = pearsonr(df['SleepHours'], df['StressLevel'])
spearman_corr, spearman_p = spearmanr(df['SleepHours'], df['StressLevel'])

print(f'Pearson correlation: {pearson_corr}, p-value: {pearson_p}')
print(f'Spearman correlation: {spearman_corr}, p-value: {spearman_p}')
```

```python
# Confirm the correlation between Gender and Cholesterol using the chi-square test
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Gender'], df['Cholesterol'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-square statistic: {chi2}, p-value: {p}')
```

### 7. Associations with Heart Disease

```python
# Confirm if there is a relationship between BMI and Heart Disease using a t-test
contingency_table = pd.crosstab(df['BMI'], df['HeartDisease'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2}, p-value: {p}')
```

### 4. Which numerical variables show the greatest dispersion (difference between Q3 and Q1)?
> - Tip: use `df.describe()` or calculate the IQR (Interquartile Range).
> - Explain why high dispersion can be relevant in a price analysis.

```python
# Which numerical variables show the greatest dispersion (difference between Q3 and Q1)?
numeric_cols = df.select_dtypes(include=[np.number]).columns
dispersion = {}
for col in numeric_cols:
    if col == 'id':  # ignore id column
        continue
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    # Check if IQR is nan
    if IQR:
        dispersion[col] = IQR

# # Sort by dispersion
# dispersion = dict(sorted(dispersion.items(), key=lambda item: item[1], reverse=True))
# print("Dispersion (IQR) of numerical variables:")
# for col, iqr in dispersion.items():
#     print(f"{col}: {iqr}")

# Print proportion of dispersion relative to the mean of each column
print("\nProportion of dispersion to mean:")
for col, iqr in dispersion.items():
    mean = df[col].mean()
    proportion = iqr / mean if mean != 0 else 0
    print(f"{col}: {proportion:.2f}")
```

```python
# Show the histogram of the 5 variables with the greatest dispersion
top_5_disp = list(dispersion.keys())[:5]
for col in top_5_disp:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Print the value range of the "bathrooms" variable, which is the most dispersed, using describe
df['bathrooms'].describe()
```

```python
# IQR means:
# The interquartile range (IQR) is a measure of statistical dispersion, indicating how spread out the values in a dataset are.
# It is calculated as the difference between the third quartile (Q3) and the first quartile (Q1) of the data.
# A high dispersion (IQR) in price indicates a wide range of house prices, which may suggest a diverse market with different types of properties.
# This is relevant for buyers and sellers, as it reflects the variability in property values, helping them make informed decisions based on market trends and property characteristics.
```

### 6. Is there any outliers in price? Show the dataframe rows with outliers.

```python
# Are there any outliers in price?
# Print the dataframe rows that contain outliers
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR))]
print("Outliers in prices:")
# print(outliers)
print(f"{len(outliers)} of {len(df)} rows are outliers in price.")
```

```python
# We used (Q1 - 1.5 * IQR) as the cutoff for lower outliers and (Q3 + 1.5 * IQR) for upper outliers.
# This means that any price below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR is considered an outlier.
# Proportion of outliers in prices
proportion_outliers = len(outliers) / len(df) * 100
print(f"Proportion of outliers in prices: {proportion_outliers:.2f}%")
```
